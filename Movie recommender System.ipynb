```python
movies['cast'][0]
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

### Search for the keyword where job: character

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'].apply(fetch_director)
```

```python
# Store it inside the crew
movies['crew'] = movies['crew'].apply(fetch_director)
```

### Steming

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():   # Splitting the string into a list based on a delimiter (e.g., space)
        y.append(ps.stem(i))  # store inside y
        
    return " ".join(y)       # convert it back into string and return it.
        
```

```python
stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron')
```

```python
new_df['tag'] = new_df['tag'].apply(stem)
```

## Text Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
# convert all the tags into vectors.
vectors = cv.fit_transform(new_df['tag']).toarray()
```

## Cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# pass all the vectors and find the cosine distance between each vector.
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity.shape
```

```python
# This is the cosine distance of first movie with every other movie.
# It's diagonal elements will be always 1. Since a movie's similarity with it's own is always 1.
similarity[0]
```

```python
similarity[0].shape
```

```python
# Sort the movie in decending order so that most similar one will come first
sorted(similarity[0], reverse = True)
```

```python
# Create the distance with their index.
list(enumerate(similarity[0]))
```

```python
# Sort in reverse order
sorted(list(enumerate(similarity[0])), reverse = True)
```

```python
# Sort it w.r.t similarity  and we need only 5 topmost similar movie
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x:x[1])[1:6]
```

### Find the 5 similar movies

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]  # Movie index of each movies.
    distances = similarity[movie_index] # Find the distance of each movie with every other movie
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6] # Get the top similar movie 
    
    # Now print it
    for i in movies_list:
        print(new_df.iloc[i[0]].title)   # Now fetch the movie name from the index

    
```

```python
# Find top 5 movies.
recommend('Batman Begins')
```
